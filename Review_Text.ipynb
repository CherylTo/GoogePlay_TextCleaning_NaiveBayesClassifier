```python
def parse2(soup):
    try:
        star = []
        reviews = []
        zipped = ()
        for rev in np.arange(0, 6):
            stars = soup.findAll("div", {"class", "tiny-star star-rating-non-editable-container"})
            stars_i = re.findall('\s\d+\s', str(stars))[rev]
            star.append(stars_i)
            review = soup.findAll("div", {"class", "review-text"})[rev].text
            reviews.append(review)
            zipped = zip(star, reviews)
    except:
        pass
            
    return list(zipped)
```

```python
for ele in game: 
    game_pre_link = ele.find('a').get('href')
    game_link= 'https://play.google.com' + str(game_pre_link)
    print(game_link)
    games_link.append(game_link)
```

```python
def get_reviews(url):
    html = urllib.request.urlopen(urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})).read()
    soup = BeautifulSoup(html, 'html5lib')
    sleep(3)
    return pd.Series(soup).apply(parse2)
```
